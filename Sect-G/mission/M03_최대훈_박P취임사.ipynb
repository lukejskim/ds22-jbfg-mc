```python
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
```

```python
# 이미지파일 numpy로 읽기
shape_img = np.array(Image.open('images/korea2.jpg'))
cloud_img = ImageColorGenerator(shape_img)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(shape_img)
plt.axis('off')
```
