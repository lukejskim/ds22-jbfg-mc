```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'http://www.kcmi.re.kr/report/report_view?report_no=1490&s_report_subject=&s_report_type=002002002013&thispage=1'
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
soup.title
```

```python
institute = soup.find('div', 're_desc')
institute
```

```python
len(institute)
```

```python
institute_txt = institute.get_text()
institute_txt
```

```python
institute_txt = institute_txt.replace('\n', ' ')
institute_txt = institute_txt.replace('\r', ' ')
institute_txt = institute_txt.replace('\t', ' ')
institute_txt = institute_txt.replace('\xa0', ' ')
institute_txt = institute_txt.strip()
institute_txt
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
institute_nouns = okt.nouns(institute_txt)
len(institute_nouns)
```

```python
institute_nouns[-10:]
```

```python
one_words = []
tow_words = []
etc_words = []

for noun in institute_nouns:
    if len(noun) == 1:
        one_words.append(noun)
    elif len(noun) == 2:
        tow_words.append(noun)
    else:
        etc_words.append(noun)
        
len(one_words), len(tow_words), len(etc_words)
```

```python
one_words = list(set(one_words))
tow_words = list(set(tow_words))
etc_words = list(set(etc_words))

len(one_words), len(tow_words), len(etc_words)
```

```python
for word in one_words:
    print(word, end='\t')
```

```python
stop_words = []
stop_words.extend(one_words)
len(stop_words)
```

```python
institute_nouns = [ word for word in institute_nouns if word not in stop_words ]
len(institute_nouns)
```

```python
institute_nouns[:10]
```

```python
import nltk
```

```python
institute = nltk.Text(institute_nouns, name='연준금리영향')
type(institute), len(institute)
```

```python
institute.vocab().most_common(30)
```

```python
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
```

```python
# 이미지파일 numpy로 읽기
shape_img = np.array(Image.open('images/interest.jfif'))
cloud_img = ImageColorGenerator(shape_img)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(shape_img)
plt.axis('off')
```

```python
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                     relative_scaling = 0.2,
                     mask = shape_img, 
                     background_color = 'white',
                     min_font_size=1, max_font_size=40).generate_from_frequencies(dict_data)
```

```python
save_img = 'wordcloud/kor_institute.png'
plt.figure(figsize=(12, 12))
plt.imshow(wordcloud.recolor(color_func=cloud_img))
plt.axis('off')
plt.savefig(save_img)
```
