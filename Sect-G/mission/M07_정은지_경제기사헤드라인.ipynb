```python
url = "https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=101"
html= requests.get(url, headers={'User-Agent':'Mozilla/5.0'}) 
Object = BeautifulSoup(html.text, "html.parser")

print(Object) 
```

```python
list = ""

for title in Object.find_all('span',"cluster_head_sub_topic"):
    list += (title.string) + " "

for title in Object.find_all('a',"cluster_text_headline nclicks(cls_sci.clsart)"):
    list += (title.string) + " "

print(list)
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
# 트위터 형태소 분석 후 명사만 추출
constitution_nouns = okt.nouns(list)
len(constitution_nouns)
```

```python
import nltk
```

```python
constitution  = nltk.Text(constitution_nouns, name='경제')
type(constitution), len(constitution)
```

```python
# 상위 30개 수 
constitution.vocab().most_common(100)
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
data = constitution.vocab().most_common(100)
len(data)
```

```python
# data를 딕셔너리 형태로 변환
dict_data = dict(data)
len(dict_data)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                      background_color='white'
                     ,relative_scaling =0.2).generate_from_frequencies(dict_data)
```

```python
plt.figure(figsize=(20,12))
plt.imshow(wordcloud)
plt.axis('off')
```
