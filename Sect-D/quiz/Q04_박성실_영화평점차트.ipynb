```python
# 막대 차트

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/MALGUN.TTF").get_name()
rc('font', family=font_name)

X = ['범죄도시2','씽2게더','귀멸의칼날','주술회전','해피투게더',
     '코다','중경상림','화양연화','아메리칸셰프','배드가이즈']
X = np.arange(10)

data = [[9.05,8.82,9.60,9.03,9.36,8.81,9.09,9.15,8.73,8.17],  #관람객
        [6.56,6.20,6.00,6.00,8.00,6.33,0.00,0.00,6.25,6.67],  #평론가
        [9.52,9.39,9.27,9.25,9.19,9.13,8.90,8.81,8.80,8.67]]  #네티즌

plt.figure(figsize=(15,6))
plt.bar(X + 0.00, data[0], color='b', width=0.25, label='관람객')
plt.bar(X + 0.25, data[1], color='g', width=0.25, label='평론가')
plt.bar(X + 0.50, data[2], color='r', width=0.25, label='네티즌')

# plt.xticks(t, ('1Q','2Q','3Q','4Q','1Q','2Q','3Q','4Q','1Q','2Q'))
# plt.xticks(X+0.2, list(X))    # set the locations of the xticks
plt.xticks(X, ('범죄도시2','씽2게더','귀멸의칼날','주술회전','해피투게더',
     '코다','중경상림','화양연화','아메리칸셰프','배드가이즈'))    # set the locations of the xticks

plt.grid()
plt.xlabel('영화 제목')
plt.ylabel('영화 평점')
plt.title('NAVER 영화 평점')
plt.legend()

plt.show()
```
