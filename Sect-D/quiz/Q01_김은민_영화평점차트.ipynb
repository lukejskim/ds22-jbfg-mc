```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc
import seaborn as sns

font_path = 'C:/Windows/Fonts/malgunsl.TTF'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
data = { 'movie' : ['범죄도시2','닥터스트레인지2','그대가조국','배드가이즈','몬스터싱어','안녕하세요',
                    '피는물보다진하다','오마주', '아치의노래', '카시오페아'],
         '관람객' : [8.55, 7.78, 9.87, 8.17, 8.25, 8.22, 5.50, 7.89, 9.42, 9.50],
         '평론가' : [6.56, 6.80, 5.00, 6.67, 6.67, 6.00, 3.00, 7.00, 6.50, 6.50],
         '네티즌' : [9.52, 7.32, 7.27, 8.67, 8.62, 8.81, 7.04, 8.33, 9.42, 9.41]}
data = pd.DataFrame(data)
```

```python
data['평균평점'] =(data['관람객'] + data['평론가'] + data['네티즌'])/3
```

```python
movie_name = data.iloc[:,0]
y1 = data.iloc[:,1] 
y2 = data.iloc[:,2] 
y3 = data.iloc[:,3] 

y4 =  data.iloc[:,4] 
```

```python
plt.figure(figsize=(20,8))
X = np.arange(len(data))  #10개의 영화

plt.bar(X+0.00, y1, color='r',width=0.25, label = '관람객',alpha=0.4)
plt.bar(X+0.30, y2, color='g',width=0.25, label = '평론가',alpha=0.5)
plt.bar(X+0.60, y3, color='b',width=0.25, label = '네티즌',alpha=0.5)

plt.grid()
plt.legend(fontsize=12)   
plt.title('Naver 영화 평점', fontsize=17)
plt.xlabel('영화 제목',fontsize=15)
plt.ylabel('영화 평점',fontsize=15)
plt.xticks(X+0.3,movie_name, fontsize=13)

plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(y4, labels= movie_name)
plt.show()
```
