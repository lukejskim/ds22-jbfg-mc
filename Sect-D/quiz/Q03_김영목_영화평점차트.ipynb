```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/MALGUN.TTF").get_name()
rc('font', family=font_name)

move_list = ['범죄의도시2', '닥터스트레인지','그대가조국','배드가이즈','몬스터싱어'
             ,'안녕하세요','피는물보다진하다','오마주','아치의노래정태춘','보드랍게']

data = [[9.05, 7.78, 9.87, 8.17,8.25,8.22,5.50,7.89,9.42,8.67],#관람객평점
        [6.56, 6.80, 5.00, 6.67,6.67,6.00,3.00,7.00,6.50,6.40],#평론가평점
        [9.52, 7.32, 7.27, 8.67,8.62,8.81,7.04,8.33,9.42,9.50]] #네티즌평점

gan = data[0]

explode = (0.2, 0, 0, 0,0,0,0,0,0,0)  
# only "explode" the 2nd slice (i.e. 'Hogs')

plt.figure(figsize=(15,15))

fig1, ax1 = plt.subplots()

ax1.pie(gan, explode=explode, labels=move_list, autopct='%1.1f%%',
        shadow=True, startangle=300, textprops={'fontsize': 10})
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
