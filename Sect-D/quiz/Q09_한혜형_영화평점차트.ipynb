```python
# 바챠트
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('my_data.txt') #텍스트파일 읽어오기

X = np.arange(5)
plt.figure(figsize=(8,5))
plt.bar(X + 0.00, data[0], color='b', width=0.25, label='cus')
plt.bar(X + 0.25, data[1], color='r', width=0.25, label='pro')
plt.bar(X + 0.50,data[2], color='y', width=0.25, label='int')


plt.xlabel('mpvie name')
plt.ylabel('score')
# plt.legend(loc= 'upper right')
plt.legend(loc= 'best')

plt.xticks(X+0.4, ('crime','doc','you','bad','monster'))
plt.title("Naver Movie Score")
plt.show()
```

```python
# 바챠트 : seaborn

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('my_data.txt') #텍스트파일 읽어오기

X = np.arange(5)
sns.bar(X + 0.00, data[0], color='b', width=0.25, label='cus')
sns.bar(X + 0.25, data[1], color='r', width=0.25, label='pro')
sns.bar(X + 0.50, data[2], color='y', width=0.25, label='int')


sns.xlabel('mpvie')
sns.ylabel('score')
# plt.legend(loc= 'upper right')
sns.legend(loc= 'best')

sns.xticks(X+0.4, ('crime','doc','you','bad','monster'))   
sns.show()
```

```python
# 파이챠트1

label = ['crime','doc','you','bad','monster']

plt.figure(figsize=(6,6))
plt.pie(y, labels=label)
plt.show()
```

```python
# 파이챠트2

labels = ['crime','doc','you','bad','monster']
explode = (0, 0.2, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')


fig1, ax1 = plt.subplots()
ax1.pie(y, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.show()
```
