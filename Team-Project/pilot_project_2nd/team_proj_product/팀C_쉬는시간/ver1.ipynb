### step3) 연도,분기 컬럼 생성
> * 컬럼명 : quarter

```python
# 연도 : str(df.brth_dt.values[0])[0:4]
# 월  : str(df.brth_dt.values[0])[4:6]
```

```python
def make_quarter(col_names):
    
    quarter = []
    
    for idx in range(len(col_names)):
        
        year = str(col_names[idx])[0:4]
        month = str(col_names[idx])[4:6]
        month = int(month) 

        if month <4:
            value = year + '.1Q'
            quarter.append(value)
        elif 4 <= month <7:
            value = year+ '.2Q'
            quarter.append(value)
        elif 7 <= month <10:
            value = year+ '.3Q'
            quarter.append(value)
        else:
            value = year+ '.4Q'
            quarter.append(value)
    
    return quarter
```

```python
df['quarter'] = make_quarter(df.exec_dt)
```
