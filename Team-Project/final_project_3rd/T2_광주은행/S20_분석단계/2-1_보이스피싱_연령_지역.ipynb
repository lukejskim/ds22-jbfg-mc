```python
voice_age_pct_df = pd.read_excel(voice_xls, sheet_name='연령별 현황_')
voice_age_pct_df.head()
```

```python
voice_age_pct_df.T
```

```python
age_20 = voice_age_pct_df.T.iloc[1]
age_30 = voice_age_pct_df.T.iloc[2]
age_40 = voice_age_pct_df.T.iloc[3]
age_50 = voice_age_pct_df.T.iloc[4]
age_60 = voice_age_pct_df.T.iloc[5]
age_70 = voice_age_pct_df.T.iloc[6]
age_range = ['2016년','2017년','2018년','2019년','2020년']
age_70
labels=age_range
plt.figure(figsize = (12,8))
plt.plot(age_range, age_20, label='20대', lw=3)
plt.plot(age_range, age_30, label='30대', lw=3)
plt.plot(age_range, age_40, label='40대', lw=3)
plt.plot(age_range, age_50, label='50대', lw=3)
plt.plot(age_range, age_60, label='60대', lw=3)
plt.plot(age_range, age_70, label='70대', lw=3)
plt.xlabel('연도')
plt.ylabel('비율(%)')
plt.grid()
plt.legend(loc=2, fontsize=13)
plt.title("\n연령별 보이스피싱 피해 현황\n", fontsize=18)
plt.show()
```

```python
age_range = ['20대 이하','30대','40대','50대','60대','70대 이상']
age_5year_avr = voice_age_df.iloc[5][2:8]
age_5year_avr
plt.figure(figsize=(6,6))
explode = [0.1,0,0,0,0.1,0.1]
plt.pie(age_5year_avr, labels=age_range, autopct='%1.1f%%',shadow=True, explode = explode, startangle=0)  
plt.title("\n연령별 보이스피싱 피해 현황\n", fontsize=13)
# plt.legend()
plt.show()
```
